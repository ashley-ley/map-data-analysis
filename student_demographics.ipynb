```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Create the path to load the file
demo_path = Path("Resources/student_demographics.csv")

# Store the data into a dataframe
demo_df = pd.read_csv(demo_path, encoding="ISO-8859-1")
demo_df.head()
```

```python
# Writing out full ethnicity in demo_df
demo_df['ethnicity'] = demo_df['ethnicity'].replace({'A': 'Asian', 'H': 'Hispanic','W':'White','AA':'African American','T': 'Arabic'})

# Writing out socioeconmic status in demo_df
demo_df['ed'] = demo_df['ed'].replace({'F': 'Free Lunch', 'R': 'Reduced Price Lunch','NaN': 'Not Applicable'})

# Changing column names to write out acronymns
demo_df = demo_df.rename(columns={'name': 'Name', 'gender': 'Sex', 'teacher': 'Teacher', 'ethnicity': 'Ethinicity', 'sped': 'Special Education','lep':'Learning English Profiency','ed':'Economically Disadvantaged', 'gt': 'Gifted and Talented', 'dyslexia': 'Dyslexia'})
demo_df.head()
```

```python
# Creating a Pie Chart for teachers
demo_df= pd.DataFrame(demo_df)

# Plotting pie chart
plt.figure(figsize=(7, 7))
demo_df['Teacher'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('2022-2023 Students per Teacher')
plt.show()
```

```python
# Creating a Pie Chart for ethnicities
demo_df= pd.DataFrame(demo_df)

# Plotting pie chart
plt.figure(figsize=(7, 7))
demo_df['Ethinicity'].value_counts(dropna=False).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('2022-2023 Student Ethnicities')
plt.show()
```

```python
# Creating a Pie Chart for socioeconomic status
demo_df= pd.DataFrame(demo_df)

# Plotting pie chart
plt.figure(figsize=(7, 7))
demo_df['Economically Disadvantaged'].value_counts(dropna=False).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('2022-2023 Student Socioeconomic Status')
plt.show()
```
