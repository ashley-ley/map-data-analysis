```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create the path to load the file
map_data_path = Path("Resources/merged_data.csv")

# Store the data into a dataframe
map_df = pd.read_csv(map_data_path, encoding="ISO-8859-1")
map_df.head()
```

```python
# Check the column names 
map_df.columns
```

```python
# Rename all the columns to describe what the values are about 
renamed_columns = map_df.rename(columns={"fa_16":"fall_2016",
                                        "wi_17":"winter_2017",
                                        "sp_17":"spring_2017",
                                        "wi_18":"winter_2018",
                                        "sp_18":"spring_2018",
                                        "wi_19":"winter_2019",
                                        "sp_19":"spring_2019",
                                        "fa_19":"fall_2019",
                                        "wi_20":"winter_2020",
                                        "sp_20":"spring_2020",
                                        "fa_20":"fall_2020",
                                        "wi_21":"winter_2021",
                                        "sp_21":"spring_2021",
                                        "fa_21":"fall_2021",
                                        "wi_22":"winter_2022",
                                        "sp_22":"spring_2022",
                                        "fa_22":"fall_2022",
                                        "wi_23":"winter_2023",
                                        "wi_23":"winter_2023",
                                        "sp_23":"spring_2023"});

map_df = renamed_columns
map_df.head()
```

```python
# Find the average MAP score grouped by each sixth grade math teacher

fall_2022_avg_by_teacher = map_df.groupby(["teacher"])["fall_2022"].mean()
fall_2022_avg_by_teacher.head()
```

```python
spring_2023_avg_by_teacher = map_df.groupby(["teacher"])["spring_2023"].mean()
spring_2023_avg_by_teacher.head()
```
