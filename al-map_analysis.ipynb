```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Create the path to load the file
map_data_path = Path("Resources/merged_data.csv")

# Store the data into a dataframe
map_df = pd.read_csv(map_data_path, encoding="ISO-8859-1")
map_df.head()
```

```python
# Check the column names 
map_df.columns
```

```python
# Rename all the columns to describe what the values are about 
renamed_columns = map_df.rename(columns={"fa_16":"fall_2016",
                                        "wi_17":"winter_2017",
                                        "sp_17":"spring_2017",
                                        "wi_18":"winter_2018",
                                        "sp_18":"spring_2018",
                                        "wi_19":"winter_2019",
                                        "sp_19":"spring_2019",
                                        "fa_19":"fall_2019",
                                        "wi_20":"winter_2020",
                                        "sp_20":"spring_2020",
                                        "fa_20":"fall_2020",
                                        "wi_21":"winter_2021",
                                        "sp_21":"spring_2021",
                                        "fa_21":"fall_2021",
                                        "wi_22":"winter_2022",
                                        "sp_22":"spring_2022",
                                        "fa_22":"fall_2022",
                                        "wi_23":"winter_2023",
                                        "wi_23":"winter_2023",
                                        "sp_23":"spring_2023"});

map_df = renamed_columns
map_df.head()
```

```python
# Rename all the columns to describe what the values are about 
renamed_columns = map_df.rename(columns={"fall_2016":"fall_2016_scores",
                                        "winter_2017":"winter_2017_scores",
                                        "spring_2017":"spring_2017_scores",
                                        "winter_2018":"winter_2018_scores",
                                        "spring_2018":"spring_2018_scores",
                                        "winter_2019":"winter_2019_scores",
                                        "spring_2019":"spring_2019_scores",
                                        "fall_2019":"fall_2019_scores",
                                        "winter_2020":"winter_2020_scores",
                                        "spring_2020":"spring_2020_scores",
                                        "fall_2020":"fall_2020_scores",
                                        "winter_2021":"winter_2021_scores",
                                        "spring_2021":"spring_2021_scores",
                                        "fall_2021":"fall_2021_scores",
                                        "winter_2022":"winter_2022_scores",
                                        "spring_2022":"spring_2022_scores",
                                        "fall_2022":"fall_2022_scores",
                                        "winter_2023":"winter_2023_scores",
                                        "winter_2023":"winter_2023_scores",
                                        "spring_2023":"spring_2023_scores"});

map_df = renamed_columns
map_df.head()
```

```python
school_year_22_23 = spring_2023_avg_by_teacher = map_df.groupby(["teacher"])["fall_2022_scores","winter_2023_scores","spring_2023_scores"].mean()
school_year_22_23.head()

school_year_22_23_df = pd.DataFrame(school_year_22_23)
school_year_22_23_df.head()
```

```python
school_year_22_23_df.plot(kind="bar",figsize=(10,6))
plt.title("Average MAP Scores by Teacher for the 2022-2023 School Year")
plt.xlabel("6th Grade Math Teacher")
plt.ylabel("Average Math MAP Scores")
plt.legend(["Fall 2022", "Winter 2023", "Spring 2023"])
plt.show()
```

```python
spring_2023_avg_by_teacher = map_df.groupby(["teacher"])["spring_2023_scores"].mean()
spring_2023_avg_by_teacher.head()
```

```python
avg_scores = map_df.groupby(["teacher"])["fall_2016_scores",
                                        "winter_2017_scores",
                                        "spring_2017_scores",
                                        "winter_2018_scores",
                                        "spring_2018_scores",
                                        "winter_2019_scores",
                                        "spring_2019_scores",
                                        "fall_2019_scores",
                                        "winter_2020_scores",
                                        "spring_2020_scores",
                                        "fall_2020_scores",
                                        "winter_2021_scores",
                                        "spring_2021_scores",
                                        "fall_2021_scores",
                                        "winter_2022_scores",
                                        "spring_2022_scores",
                                        "fall_2022_scores",
                                        "winter_2023_scores",
                                        "fall_2022_scores",
                                        "winter_2023_scores",
                                        "spring_2023_scores"].mean()
avg_scores.head()

avg_scores_df = pd.DataFrame(avg_scores)
avg_scores_df.head()
```

```python
# Transpose the DataFrame
avg_scores_df = avg_scores_df.transpose()

# Plot the line graph
avg_scores_df.plot(kind="line",figsize=(10,6))
plt.title("Average MAP Scores by Teacher per Year")
plt.xlabel("Years")
plt.ylabel("Average Math MAP Scores")
plt.legend(title="Teacher", loc="upper left")
plt.show()
```
